```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data = {
    "ArtifactID": np.arange(1, 101),
    "LocationCode": np.random.choice(["A1", "B2", "C3", "D4"], 100),
    "ValueCode": np.random.randint(10, 100, 100),
    "Era": np.random.choice(["Ancient", "Medieval", "Renaissance"], 100),
    "DiscoveryYear": np.random.randint(1800, 2023, 100),
    "EncodedDescription": ["Xyz" + str(i) for i in np.random.randint(100, 1000, 100)]
}

eldoria_df = pd.DataFrame(data)

# Decoding the 'ValueCode' (Example: Multiply by 10)
eldoria_df["ArtifactValue"] = eldoria_df["ValueCode"] * 10

# Decoding the 'LocationCode' (Example: Adding a region name)
location_mapping = {"A1": "North Region", "B2": "South Region", "C3": "East Region", "D4": "West Region"}
eldoria_df["Location"] = eldoria_df["LocationCode"].map(location_mapping)
```

```python
avg_era_value=eldoria_df.groupby("Era")["ArtifactValue"].mean()
# print(eldoria_df.groupby("Era")["ArtifactValue"].mean())
print(avg_era_value)
```

```python
discovery_year_counts=eldoria_df["DiscoveryYear"].value_counts()
# print("discovery_year_counts:",discovery_year_counts)
max_discovery_year=discovery_year_counts.idxmax()
year_3_count=discovery_year_counts[discovery_year_counts==3].index.tolist()
print(year_3_count)
# print("max_discovery_year:",max_discovery_year)
```
