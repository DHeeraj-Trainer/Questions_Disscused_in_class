# Python Data Visualization Cheat Sheet ‚Äì Matplotlib & Seaborn
This cheat sheet covers **Matplotlib** and **Seaborn** for **data visualization**, including major plots, syntax, explanations, and color palettes.

---

## üîπ Matplotlib ‚Äì Basic Plots

### **1Ô∏è‚É£ Line Plot üìà**
üîπ **Used for** visualizing trends over time.

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 20, 25, 30, 40]

plt.plot(x, y, marker='o', linestyle='-', color='m', label='Trend')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot Example')
plt.legend()
plt.grid(True)
plt.show()
```

### **2Ô∏è‚É£ Bar Plot üìä**
üîπ **Used for** comparing categorical data.

```python
categories = ['A', 'B', 'C', 'D']
values = [10, 25, 17, 30]

plt.bar(categories, values, color='skyblue', edgecolor='black')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Plot Example')
plt.show()
```

### **3Ô∏è‚É£ Scatter Plot üîµ**
üîπ **Used for** visualizing relationships between variables.

```python
import numpy as np

x = np.random.rand(50)
y = np.random.rand(50)

plt.scatter(x, y, color='red', marker='x')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot Example')
plt.show()
```

### **4Ô∏è‚É£ Histogram üìä**
üîπ **Used for** showing frequency distributions.

```python
data = np.random.randn(1000)

plt.hist(data, bins=20, color='blue', edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram Example')
plt.show()
```

### **5Ô∏è‚É£ Pie Chart üçï**
üîπ **Used for** displaying proportions.

```python
labels = ['A', 'B', 'C', 'D']
sizes = [25, 35, 20, 20]
colors = ['gold', 'lightblue', 'lightgreen', 'pink']

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.title('Pie Chart Example')
plt.show()
```

### **6Ô∏è‚É£ Box Plot üì¶**
üîπ **Used for** showing data distribution and detecting outliers.

```python
data = [np.random.randn(100), np.random.randn(100) + 2, np.random.randn(100) - 2]

plt.boxplot(data, labels=['A', 'B', 'C'])
plt.xlabel('Category')
plt.ylabel('Values')
plt.title('Box Plot Example')
plt.show()
```

## üîπ Seaborn ‚Äì Advanced Plots

### **1Ô∏è‚É£ Seaborn Line Plot**
üîπ **Used for** trends with confidence intervals.

```python
import seaborn as sns
import pandas as pd

data = pd.DataFrame({'x': range(10), 'y': [i + (i%3) for i in range(10)]})
sns.lineplot(x='x', y='y', data=data, marker='o')

plt.title('Seaborn Line Plot')
plt.show()
```

### **2Ô∏è‚É£ Seaborn Bar Plot**
üîπ **Used for** categorical comparisons.

```python
sns.barplot(x=['A', 'B', 'C', 'D'], y=[10, 25, 17, 30], palette='coolwarm')
plt.title('Seaborn Bar Plot')
plt.show()
```

### **3Ô∏è‚É£ Seaborn Heatmap üî•**
üîπ **Used for** showing relationships in matrices.

```python
data = np.random.rand(5, 5)
sns.heatmap(data, annot=True, cmap='coolwarm')
plt.title('Seaborn Heatmap')
plt.show()
```

| **Pie Chart** | `plt.pie()` | (Matplotlib only) |
| **Box Plot** | `plt.boxplot()` | `sns.boxplot()` |
| **Heatmap** | `sns.heatmap()` | `sns.heatmap()` |
