````python
import numpy as np

encoded_map = np.array([
    [1, 0, 2, 1, 0],
    [0, 2, 1, 2, 1],
    [1, 1, 0, 0, 2],
    [2, 0, 1, 2, 0],
    [0, 2, 2, 1, 1]
])


moves={
    0:(0,1),
    1:(1,0),
    2:(1,1),
}

x,y,path=0,0,[(0,0)]

while (x<5 and y<5):
    dx,dy=moves[encoded_map[x,y]]
    if x+dx>=5 or y+dy>=5:
        break
    x,y=x+dx,y+dy
    path.append((x,y))


print("Final Destination: ",(x,y))
print("Path: ",path)


````

3.  Follow the clues.
4.  Treasure: Final position.
5.  Ignore moves outside the grid.

Your Tasks:

1.  Decode the clues (loops, conditionals).
2.  Track the path (list of coordinates).
3.  Report the treasure's location.
4.  Show the journey's path.

Questions:

1.  Treasure coordinates?
2.  Journey's path?
3.  If 4=Up-Left, would they leave the grid?

Share your Python code and answers below! üêçüí∞

````python
import numpy as np

clue_grid = np.array([
    [2, 1, 0, 3, 1],
    [1, 3, 2, 1, 0],
    [0, 2, 3, 0, 2],
    [3, 1, 0, 2, 3],
    [1, 0, 2, 3, 1]
])

row, col = 2, 2
path = [(row, col)]

while True:
    instruction = clue_grid[row, col]

    if instruction == 0:
        if row - 1 >= 0:
            row -= 1
    elif instruction == 1:
        if row + 1 < clue_grid.shape[0]:
            row += 1
    elif instruction == 2:
        if col - 1 >= 0:
            col -= 1
    elif instruction == 3:
        if col + 1 < clue_grid.shape[1]:
            col += 1
    else:
        break #if there was a 4, this would be where it would be processed.

    path.append((row, col))

    if row < 0 or row >= clue_grid.shape[0] or col < 0 or col >= clue_grid.shape[1]:
        break

print("1. Treasure Coordinates:", (row, col))
print("2. Journey's Path:", path)
````
