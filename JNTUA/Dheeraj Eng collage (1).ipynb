```python
import pandas as pd
import random
from faker import Faker

# Initialize Faker for fake data generation
fake = Faker()

# Function to generate random phone numbers (some incorrect)
def generate_phone():
    if random.random() < 0.1:  # 10% chance of an invalid phone number
        return fake.word()  # Return a random word instead of a number
    else:
        return fake.phone_number()

# Generate 200+ guest records
guest_data = []
for i in range(220):  
    name = fake.name()
    phone = generate_phone()
    email = fake.email()
    
    # Introduce duplicate guest entries (same name, different contact)
    if random.random() < 0.1:  # 10% chance of a duplicate guest
        guest_data.append([name, generate_phone(), fake.email()])
    
    guest_data.append([name, phone, email])

guest_df = pd.DataFrame(guest_data, columns=["guest_name", "phone", "email"])

# Generate 200+ booking records
booking_data = []
for i in range(250):
    guest_id = random.randint(1, len(guest_df))  # Random guest
    check_in = fake.date_between(start_date="-30d", end_date="+30d")
    check_out = fake.date_between(start_date=check_in, end_date="+40d")

    # Introduce incorrect check-in/check-out dates
    if random.random() < 0.1:  # 10% chance of incorrect dates
        check_in, check_out = check_out, check_in  # Swap dates
    
    room_number = random.choice(["101", "102", "103", "104", "105", "", None])  # Some missing room numbers
    amount = round(random.uniform(2000, 10000), 2)  # Random booking amount
    
    booking_data.append([guest_id, check_in, check_out, room_number, amount])

booking_df = pd.DataFrame(booking_data, columns=["guest_id", "check_in", "check_out", "room_number", "amount"])

# Generate Room Table
room_data = []
for i in range(1, 106):
    if random.random() < 0.05:  # 5% chance of missing room records
        continue
    room_type = random.choice(["Single", "Double", "Suite", None])  # Some missing room types
    price_per_night = round(random.uniform(1500, 8000), 2)

    room_data.append([i, room_type, price_per_night])

room_df = pd.DataFrame(room_data, columns=["room_number", "room_type", "price_per_night"])

# Save to Excel
with pd.ExcelWriter("hotel_messy_data.xlsx") as writer:
    guest_df.to_excel(writer, sheet_name="Guests", index=False)
    booking_df.to_excel(writer, sheet_name="Bookings", index=False)
    room_df.to_excel(writer, sheet_name="Rooms", index=False)

print("Messy hotel dataset generated successfully! ðŸŽ‰")
```
